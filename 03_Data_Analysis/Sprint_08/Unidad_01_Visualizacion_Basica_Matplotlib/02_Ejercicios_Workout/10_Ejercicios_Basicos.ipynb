### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
x = np.linspace(-2*np.pi, 2*np.pi, 300)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)          # hueco 1 (izquierda)
plt.plot(x, np.sin(x))
plt.title('Seno')

plt.subplot(1, 2, 2)          # hueco 2 (derecha)
plt.plot(x, np.cos(x))
plt.title('Coseno')

plt.tight_layout()
plt.show()
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python
x = np.linspace(-2*np.pi, 2*np.pi, 300)

fig, axes = plt.subplots(3, 1, figsize=(8, 9))

axes[0].plot(x, np.sin(x))
axes[0].set_title('Seno')

axes[1].plot(x, np.cos(x))
axes[1].set_title('Coseno')

axes[2].plot(x, np.tan(x))
axes[2].set_ylim(-5, 5)       # limita el eje Y (tan se dispara)
axes[2].set_title('Tangente')

plt.tight_layout()
plt.show()
```


### Ejercicio 3: Ajuste de Colores y Estilos de Línea  

Modifica la primera figura para que se muestre de la siguiente manera

```python
x = np.linspace(-2*np.pi, 2*np.pi, 300)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x), 'r--', linewidth=2)  # rojo discontinuo
plt.title('Seno')

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x), 'g-.', linewidth=2)  # verde punto-raya
plt.title('Coseno')

plt.tight_layout()
plt.show()

# Colores: 'r'=rojo 'g'=verde 'b'=azul 'k'=negro 'm'=magenta
# Líneas:  '-'=sólida '--'=discontinua '-.'=punto-raya ':'=punteada
```
