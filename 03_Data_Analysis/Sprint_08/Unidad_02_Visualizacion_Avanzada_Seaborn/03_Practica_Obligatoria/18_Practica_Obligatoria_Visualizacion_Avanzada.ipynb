### #1.2

Crea una función similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una función para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numérica del dataframe
3. Una columna numérica del dataframe diferente a la anterior.
4. Un columna categórica (Opcional, valor por defecto None)
5. Un valor para el tamaño de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamaño
6. Un sexto argumento "escala", que permita manipular la escala del tamaño en el caso de que el quinto argumento sea una cadena. Este argumento multiplicará a los valores de la columna indicada en el parámetro anterior (así si quiero dividir por 10000, este sexto argumento tendrá que valer 1/10000)

La función debe pintar el diagrama de dispersión de la columna numérica 1, en el eje X, frente a la columna numérica 2, en el eje Y, usando un código de colores en función de la columna categórica y con el tamaño de los puntos según el argumento correspondiente. 

Si no se diera una columna categórica como argumento, el color no debe pasarse como parámetro a la función de Seaborn o Matplotlib que uses.

EXTRA OPCIONAL: Añade un argumento que controle si la función pinta una leyenda para colores y tamaños o un colorbar y una leyenda para los tamaños.

```python

def scatter_avanzado(df, col_x, col_y, col_cat=None, size=50, escala=1, legend=True):
    """
    Diagrama de dispersión configurable.

    Parámetros
    ----------
    df      : DataFrame de pandas
    col_x   : columna numérica para eje X
    col_y   : columna numérica para eje Y
    col_cat : columna categórica para colores (None → sin hue)
    size    : número (tamaño fijo) o str (nombre de columna que controla el tamaño)
    escala  : multiplicador aplicado a los valores de la columna size
    legend  : True → leyenda discreta; False → colorbar + leyenda de tamaños
    """
    fig, ax = plt.subplots(figsize=(10, 6))

    # Calcular tamaños
    if isinstance(size, str):
        sizes = df[size] * escala
    else:
        sizes = size

    # Colores según categoría
    if col_cat is not None:
        categorias = df[col_cat].unique()
        palette = sns.color_palette("tab10", len(categorias))
        color_map = dict(zip(categorias, palette))
        colors = df[col_cat].map(color_map)

        sc = ax.scatter(df[col_x], df[col_y], c=colors, s=sizes, alpha=0.7, edgecolors='white', linewidths=0.3)

        if legend:
            handles = [plt.Line2D([0],[0], marker='o', color='w',
                                   markerfacecolor=color_map[cat], markersize=8, label=cat)
                       for cat in categorias]
            ax.legend(handles=handles, title=col_cat, bbox_to_anchor=(1.01, 1), loc='upper left')
        else:
            # Colorbar usando valores numéricos si la columna lo permite
            try:
                norm = plt.Normalize(df[col_cat].min(), df[col_cat].max())
                sm = plt.cm.ScalarMappable(cmap='tab10', norm=norm)
                plt.colorbar(sm, ax=ax, label=col_cat)
            except Exception:
                pass
    else:
        sc = ax.scatter(df[col_x], df[col_y], s=sizes, alpha=0.7, edgecolors='white', linewidths=0.3)

    ax.set_xlabel(col_x)
    ax.set_ylabel(col_y)
    ax.set_title(f'{col_y} vs {col_x}')
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```

Una vez calculado el logaritmo, también puedes probar a categorizar la población, aquí tienes un ejemplo de como podría salir para que lo tengas de referencia:

```python
def scatter_avanzado(df, col_x, col_y, col_cat=None, size=50, escala=1, legend=True):
    fig, ax = plt.subplots(figsize=(10, 6))

    # Calcular tamaños
    if isinstance(size, str):
        sizes = df[size] * escala
    else:
        sizes = size

    if col_cat is not None:
        # Convertir a string para evitar problemas con Categorical
        col_cat_vals = df[col_cat].astype(str)
        categorias = col_cat_vals.unique()
        palette = sns.color_palette("tab10", len(categorias))
        color_map = dict(zip(categorias, palette))
        colors = col_cat_vals.map(color_map)

        ax.scatter(df[col_x], df[col_y], c=list(colors), s=sizes,
                   alpha=0.7, edgecolors='white', linewidths=0.3)

        if legend:
            handles = [plt.Line2D([0], [0], marker='o', color='w',
                                  markerfacecolor=color_map[cat],
                                  markersize=8, label=cat)
                       for cat in sorted(categorias)]
            ax.legend(handles=handles, title=col_cat,
                      bbox_to_anchor=(1.01, 1), loc='upper left')
    else:
        ax.scatter(df[col_x], df[col_y], s=sizes,
                   alpha=0.7, edgecolors='white', linewidths=0.3)

    ax.set_xlabel(col_x)
    ax.set_ylabel(col_y)
    ax.set_title(f'{col_y} vs {col_x}')
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```

## #2: Presentación

En esta parte tendrás que hacer dos cosas, crear una presentación sobre el titanic de no más de 6 slides, sin incluir portada, con los siguientes mensajes y su apoyo en datos:  
1- El titanic fue un accidente con una tasa de mortalidad alta.  
2- En esa ocasión los datos apoyan que se dio el "Las mujeres y los niños primero", pero...  
3- ... que la clase en la que se viajes tuvo una influencia significativa y podría explicar...  
4- ... las diferencias de tasas de supervivencia entre mujeres y niños.    
5- Además aunque el puerto de embarque parece influir, se podría asumir que lo fue el tipo de pasaje que embarcó en cada uno.  

Y generar el notebook con las gráficas que vas a emplear en la presentación. Tienes que entregar ambas cosas.

```python
# Carga (usa seaborn si tienes conexión, o tu CSV local)
try:
    titanic = sns.load_dataset('titanic')
except Exception:
    titanic = pd.read_csv('./data/titanic.csv')

titanic = titanic.dropna(subset=['age', 'sex', 'pclass', 'survived', 'embarked'])
titanic['who'] = titanic.apply(
    lambda r: 'child' if r['age'] < 16 else ('man' if r['sex'] == 'male' else 'woman'), axis=1
)
print(titanic.shape)
titanic.head()
```

```python
total = len(titanic)
sobrevivieron = titanic['survived'].sum()
murieron = total - sobrevivieron

fig, ax = plt.subplots(figsize=(6, 5))
ax.pie(
    [sobrevivieron, murieron],
    labels=['Sobrevivieron', 'Fallecieron'],
    autopct='%1.1f%%',
    colors=['#2ecc71', '#e74c3c'],
    startangle=90,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)
ax.set_title(f'Titanic: {total} pasajeros\nTasa de mortalidad elevada', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
surv_who = titanic.groupby('who')['survived'].mean().reset_index()
surv_who.columns = ['Grupo', 'Tasa supervivencia']
surv_who['Grupo'] = surv_who['Grupo'].map({'man': 'Hombres', 'woman': 'Mujeres', 'child': 'Niños'})

fig, ax = plt.subplots(figsize=(6, 5))
sns.barplot(data=surv_who, x='Grupo', y='Tasa supervivencia',
            palette=['#3498db', '#e74c3c', '#f39c12'], ax=ax)
ax.set_ylim(0, 1)
ax.set_ylabel('Tasa de supervivencia')
ax.set_title('¿Se cumplió "Las mujeres y los niños primero"?')
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0%}',
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=11)
plt.tight_layout()
plt.show()
```

```python
surv_class = titanic.groupby('pclass')['survived'].mean().reset_index()

fig, ax = plt.subplots(figsize=(6, 5))
sns.barplot(data=surv_class, x='pclass', y='survived',
            palette=['#f1c40f', '#95a5a6', '#cd7f32'], ax=ax)
ax.set_ylim(0, 1)
ax.set_xlabel('Clase')
ax.set_ylabel('Tasa de supervivencia')
ax.set_title('La clase social marcó las probabilidades de sobrevivir')
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0%}',
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=11)
plt.tight_layout()
plt.show()
```

```python
surv_class_who = titanic.groupby(['pclass', 'who'])['survived'].mean().reset_index()
surv_class_who['who'] = surv_class_who['who'].map({'man': 'Hombres', 'woman': 'Mujeres', 'child': 'Niños'})

fig, ax = plt.subplots(figsize=(8, 5))
sns.barplot(data=surv_class_who, x='pclass', y='survived', hue='who',
            palette=['#3498db', '#e74c3c', '#f39c12'], ax=ax)
ax.set_ylim(0, 1)
ax.set_xlabel('Clase')
ax.set_ylabel('Tasa de supervivencia')
ax.set_title('La clase explica diferencias entre mujeres y niños')
ax.legend(title='Grupo')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Supervivencia por puerto
surv_emb = titanic.groupby('embarked')['survived'].mean().reset_index()
sns.barplot(data=surv_emb, x='embarked', y='survived',
            palette='Set2', ax=axes[0])
axes[0].set_ylim(0, 1)
axes[0].set_title('Supervivencia por puerto de embarque')
axes[0].set_xlabel('Puerto (C=Cherburgo, Q=Queenstown, S=Southampton)')
axes[0].set_ylabel('Tasa de supervivencia')
for p in axes[0].patches:
    axes[0].annotate(f'{p.get_height():.0%}',
                     (p.get_x() + p.get_width() / 2, p.get_height()),
                     ha='center', va='bottom', fontsize=10)

# Distribución de clases por puerto (el verdadero factor)
class_emb = titanic.groupby(['embarked', 'pclass']).size().reset_index(name='n')
class_emb_pct = class_emb.copy()
totales = class_emb_pct.groupby('embarked')['n'].transform('sum')
class_emb_pct['pct'] = class_emb_pct['n'] / totales

sns.barplot(data=class_emb_pct, x='embarked', y='pct', hue='pclass',
            palette=['#f1c40f', '#95a5a6', '#cd7f32'], ax=axes[1])
axes[1].set_title('Distribución de clases por puerto\n(el puerto ≈ tipo de pasajero)')
axes[1].set_xlabel('Puerto')
axes[1].set_ylabel('Proporción de pasajeros')
axes[1].legend(title='Clase')

plt.tight_layout()
plt.show()
```
